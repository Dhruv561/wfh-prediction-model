```python
import pandas as pd
```

```python
household = pd.read_csv("data/cleaned_household.csv")
person = pd.read_csv("data/cleaned_person.csv")

combinedData = person.merge(household, on="hhid", how="inner")
combinedData.to_csv("data/combined_data.csv", index=False)
print(f"Combined Dataset Rows: {combinedData.shape[0]}")
print(f"Rows in Data Before Merging: {max(household.shape[0], person.shape[0])}")
```
